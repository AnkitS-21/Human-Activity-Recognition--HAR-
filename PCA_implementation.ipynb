```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer= load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
cancer['feature_names'], len(cancer['feature_names'])

"""We'll try to extract 2 principal dimensions from these 30 features"""
```

```python
cancer["target"], len(cancer["target"])

"""Code{"Cancer-malignant": 0, "Cancer-benign": 1}"""
```

```python
df= pd.DataFrame(cancer["data"], columns= cancer["feature_names"])
df.head()
```

```python
"""Before we can apply PCA, we need to scale/normalise(std=1, mean=0) the data"""

from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit(df)

scaled_data= scaler.transform(df)

scaled_data.mean(axis= 0), scaled_data.std(axis= 0)
#All the columns in the scaled_dat have mean 0 and std 1
```

```python
scaled_df= pd.DataFrame(scaled_data, columns= cancer["feature_names"])
scaled_df.head()
```

```python
from sklearn.decomposition import PCA

pca= PCA(n_components= 2)
pca.fit(scaled_df)

x_pca= pca.transform(scaled_df)

print(f"Earlier shape: {scaled_df.shape}")  
print(f"Later shape: {x_pca.shape}")
```

```python
x_pca_df= pd.DataFrame(x_pca, columns= ["PC1", "PC2"])
```

```python
compressed_data= pd.concat([x_pca_df, pd.Series(cancer["target"])],axis= 1)
compressed_data.columns= ["PC1", "PC2", "target"]
compressed_data.head()
```

```python
plt.scatter(compressed_data["PC1"], compressed_data["PC2"], c= cancer["target"], cmap= "plasma")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
